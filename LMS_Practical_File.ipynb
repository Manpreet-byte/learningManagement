# Learning Management System API Practical Assignment

## Project: Learning Management System API

## 1. Objective

Build a backend API where users can register/login, create and manage courses, and enroll students in courses using secure routes.

## 2. Tools and Technologies

- Node.js
- Express.js
- MongoDB
- Mongoose
- JWT (JSON Web Token)
- BcryptJS
- Postman
- Nodemon

## 3. System Architecture Diagram (Text)

Client (Postman/Web App) -> LMS Server (Express API) -> MongoDB Database

### Explanation

- The client sends HTTP requests to the LMS API.
- The server handles authentication, business logic, and validations.
- The database stores users, courses, and enrollment records.
- Protected routes use JWT middleware before controller logic runs.

## 4. Project Folder Structure

```text
learningManagementSystem/
  config/
    db.js
  controllers/
    authController.js
    courseController.js
    enrollController.js
    studentController.js
  middleware/
    authMiddleware.js
    errorHandler.js
    logger.js
  models/
    User.js
    Course.js
    Enrollment.js
  routes/
    authRoutes.js
    courseRoutes.js
    enrollRoutes.js
    studentRoutes.js
  app.js
  server.js
```

This structure keeps code modular and easier to maintain.

## 5. Database Schema

### User Schema

- name: Student/Instructor name
- email: Unique login email
- password: Hashed password
- role: student or instructor

### Course Schema

- title: Course name
- description: Course details
- instructor: Reference to User (instructor)

### Enrollment Schema

- student: Reference to User (student)
- course: Reference to Course
- createdAt/updatedAt: Enrollment timestamps

Each schema is connected using ObjectId references to build relationships.

## 6. API Endpoints Table

| Method | Endpoint | Purpose |
|---|---|---|
| POST | /api/auth/register | Register user |
| POST | /api/auth/login | Login user and get token |
| POST | /api/courses | Create course (auth) |
| GET | /api/courses | Get all courses |
| GET | /api/courses/:id | Get one course |
| PUT | /api/courses/:id | Update course (auth) |
| DELETE | /api/courses/:id | Delete course (auth) |
| POST | /api/enroll | Create enrollment (auth) |
| GET | /api/enroll | Get enrollments (auth) |
| PUT | /api/enroll/:id | Update enrollment (auth) |
| DELETE | /api/enroll/:id | Remove enrollment (auth) |

## 7. Sample Request and Response

### Sample Request: Create Course

```http
POST /api/courses
Authorization: Bearer <lms_token>
Content-Type: application/json

{
  "title": "Web Development Basics",
  "description": "Introduction to HTML, CSS, and JavaScript"
}
```
